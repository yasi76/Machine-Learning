```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Sample data (X = years of experience, Y = salary)
X = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
Y = np.array([20000, 24000, 28000, 30000, 36000, 50000, 72000, 85000])

# Transform features to polynomial degree 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict
Y_pred = model.predict(X_poly)

# Plot data and polynomial regression curve
plt.scatter(X, Y, color="blue")
plt.plot(X, Y_pred, color="red")
plt.title("Polynomial Regression: Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Polynomial terms up to degree 2 included in transformed features
print(X_poly)
```

```python
# Predict for an new value (e.g., 9 years of experience)
X_new = np.array([9]).reshape(-1, 1)
X_new_poly = poly.transform(X_new)
Y_new_pred = model.predict(X_new_poly)

# Changed Y_unseen_pred to Y_new_pred
print(f"Predicted salary for 9 years of experience: {Y_new_pred[0]:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate some example data
np.random.seed(42)
X = np.linspace(1, 10, 100).reshape(-1, 1)  # X values

# y = 2e^(0.3x)
y = 2 * np.exp(0.3 * X).ravel() + np.random.normal(0, 1.2, 100)

# Apply natural log transformation to y
y_log = np.log(y)

# Perform linear regression on transformed data
model = LinearRegression()
model.fit(X, y_log)

# Make predictions
y_pred_log = model.predict(X)
y_pred = np.exp(y_pred_log)  # Convert back to exponential scale

# Plot results
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='Exponential fit')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Exponential Regression using LinearRegression')
plt.legend()
plt.show()

# Print the model parameters
a = np.exp(model.intercept_)
b = model.coef_[0]
print(f"Exponential regression equation: y = {a:.2f} * e^({b:.2f} * x)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate some example data
np.random.seed(42)
X = np.linspace(1, 10, 100).reshape(-1, 1)  # X values

# y = 3 + 2log(x)
y = 3 + 2 * np.log(X).ravel() + np.random.normal(0, 0.5, 100)

# Apply log transformation to X
X_log = np.log(X)

# Perform linear regression on transformed data
model = LinearRegression()
model.fit(X_log, y)

# Make predictions
y_pred = model.predict(X_log)

# Plot results
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='Logarithmic fit')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Logarithmic Regression using LinearRegression')
plt.legend()
plt.show()

# Print the model parameters
a = model.intercept_
b = model.coef_[0]
print(f"Logarithmic regression equation: y = {a:.2f} + {b:.2f} * log(x)")
```
