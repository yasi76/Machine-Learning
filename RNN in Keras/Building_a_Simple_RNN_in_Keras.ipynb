```python
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense
import numpy as np

# Generate dummy sequential data
sequence = np.arange(0, 1000, 5)
time_steps = 10
X_train = []
y_train = []
for i in range(len(sequence) - time_steps):
    X_train.append(sequence[i:i + time_steps])
    y_train.append(sequence[i + time_steps])

# Reshape data for RNN input
# The original code had x_train instead of X_train, causing the issue.
# Keras expects X_train to have shape (num_samples, time_steps, num_features)
X_train = np.array(X_train).reshape(len(X_train), time_steps, 1)
y_train = np.array(y_train).reshape(len(y_train), 1)

# Create a simple RNN model
model = Sequential()
model.add(SimpleRNN(50, input_shape=(time_steps, 1), activation='relu'))
model.add(Dense(1))  # Output layer for regression

# Show model structure
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=250, batch_size=32)
```
