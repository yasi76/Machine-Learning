```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np

# Load the Iris dataset
data = load_iris()
X = data.data
y = data.target

# Standardize the data
X = (X - X.mean(axis=0)) / X.std(axis=0)

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X)

# Get the explained variance ratios
explained_variance_ratios = pca.explained_variance_ratio_

# Plot the percentages of explained variance as a bar graph
plt.bar(np.arange(1, len(explained_variance_ratios) + 1), explained_variance_ratios)
plt.xlabel("Principal Component")
plt.ylabel("Percentage of Explained Variance")
plt.title("Explained Variance Ratio per Principal Component")
plt.xticks(np.arange(1, len(explained_variance_ratios) + 1))
plt.show()
```

```python
# Plot the PCA-transformed data for the first two components
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.title("PCA of Iris Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
